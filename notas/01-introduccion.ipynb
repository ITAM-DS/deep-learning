# Introducción

<h2>1.1 Inteligencia artificial</h2>

Este concepto empieza a generar interes alrededor de los años 50's (<a href="https://en.wikipedia.org/wiki/Alan_Turing">Alan Turing</a> e <a href="https://en.wikipedia.org/wiki/Isaac_Asimov">Isaac Asimov</a>). 

<h5>Definición</h5>

El esfuerzo de automatizar tareas que son de naturaleza intelectual, usualmente realizadas por personas, para que las realice una computadora. Es decir se quieren emular pcoesos cognitivos de una persona.

  <p style="padding-left:10px;">&#9432;  Existen varias tareas que no forzosamente requieren de un procedimiento de aprendizaje.
</p>
</div>

<h3>Inteligencia basada en reglas (IA simbólica)</h3>

Si escribimos todas las reglas posibles, entonces somos capaces de manipular conceptos como si lo hiciera un agente inteligente. Por ejemplo: los sistemas expertos. Este concepto comenzó a escucharse entre los 50's y 80's.

<h2>1.2 Aprendizaje automático</h2>

Es una rama de la inteligencia artificial que permite que las máquinas aprendan sin ser expresamente programadas para ello. Este término comienza a ser escuchado alrededor de los años 90's y principalmente se basan en prueba y error (empíricos). Estos modelos han evolucionado favorablemente debido a que cada vez existe más capacidad de cómputo (procesamiento) y a los grandes volúmenes de datos (almacenamiento). 

Un algoritmo recibe un conjunto de reglas y datos para producir una respuesta.

<h2>1.3 Aprendizaje profundo</h2>

El <a href="https://www.bbvaopenmind.com/tecnologia/mundo-digital/que-es-el-aprendijaze-profundo/">aprendizaje profundo</a> implica la creación de redes neuronales a gran escala que permiten que el ordenador aprenda y "piense" por sí mismo sin necesidad de intervención humana directa. Consiste en varias capasque se aprenden al mismo tiempo, a diferencia de aprendizaje automático que intenta encontrar la mejor optimización <i>h</i>.

<center><img src="01-images/dl.jpg" width="350px" height="350px"/></center>

<p>
<div style="padding-top:15px;padding-bottom:2px;background-color: #e7f3fe;border-left: 6px solid #2196F3;">

</p>
</div>
<br>

Entre las aplicaciones de aprendizaje profundo encontramos:
- Clasificación de imágenes (e.g. <a href="http://www.image-net.org/">Imagenet</a>) 
- Segmentación de imágenes
- Transferencia de estilos artísticos
- Modelos generativos
- Descripciones de imágenes
- Traducción de lenguaje
- Síntesis de texto
- Preguntas y respuestas (texto e imágenes)
- Juegos de estrategia

<h5>Definición</h5>

Representación: Una transformación de datos que hace una tarea más sencilla.

<h3>Diferencias</h3>

<h4>Aprendizaje automático</h4>

Su función se ve expresada de la siguiente forma:

$$\hat{y_n} = h(x) \approx y $$

Con una función de pérdida:

$$ 
\ell(\hat{y_n}, y)
$$
$$
n \in H_{ML}
$$
$$
h^{*} = argmin(\ell(\hat{y_n}, y))
$$


<h4>Aprendizaje profundo</h4>

Representación en capas expresada de las siguiente forma:

$$
\hat{y_n} = h_m(\cdots(h_2(h_1(x)))) \approx y
$$

Con una función de pérdida:

$$
\ell(\hat{y_n}, y)
$$
$$
h = h_1 \times h_2 \times \cdots \times h_m
$$
$$
n \in H_{DL}
$$

<center><h1>¡No!</h1></center>

<ul>
